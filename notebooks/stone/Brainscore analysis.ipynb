```python
def scatter(k1, k2, ignore=[]):
    fig, ax = plt.subplots(figsize=(10,6))
    X, Y = [], []
    k2s = k2.split(".")
    for k, v in data.items():
        if k in ignore: continue
        # k is key name
        txt = k
        y = v
        for b in k2s:
            y = y[b]
        x = v[k1]
        X.append(x)
        Y.append(y)
        ax.annotate(txt, (x, y))
    plt.xlabel(k1)
    plt.ylabel(k2)
    plt.tight_layout()
    ax.scatter(X, Y)
    #     
```
